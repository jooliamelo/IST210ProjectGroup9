```
import sqlite3

#Connect to database
connectionToDatabase = sqlite3.connect('GoogleShopping') 
```

```
#Drop Customers table if there already is one
cursor = connectionToDatabase.cursor()
cursor.execute('''drop table IF EXISTS Customers''')
connectionToDatabase.commit()

#Create Customers table
cursor.execute('''create table Customers (customerID PRIMARY KEY, Name, Address, Payment, Email)''')
connectionToDatabase.commit()
```

```
#Drop Transactions table if there already is one
cursor = connectionToDatabase.cursor()
cursor.execute('''drop table IF EXISTS Transactions''')
connectionToDatabase.commit()

#Create Transactions table
cursor.execute('''create table Transactions (TransactionID PRIMARY KEY, CustomerID FORIEGN KEY, ArrivalDate, Barcode FORIEGN KEY, TransactionDate)''')
connectionToDatabase.commit()
```

```
#Drop WebsiteInformation table if there already is one
cursor = connectionToDatabase.cursor()
cursor.execute('''drop table IF EXISTS WebsiteInformation''')
connectionToDatabase.commit()

#Create WebsiteInformation table
cursor.execute('''create table WebsiteInformation (URL PRIMARY KEY, Name, TotalSales Int, CommissionRate, TotalCommission)''')
connectionToDatabase.commit()
```

```
#Drop Items table if there already is one
cursor = connectionToDatabase.cursor()
cursor.execute('''drop table IF EXISTS Items''')
connectionToDatabase.commit()

#Create Item table
cursor.execute('''create table Items (Barcode PRIMARY KEY, ItemPrice, OriginalSeller, ShippingPrice, ItemType, URL FORIEGN KEY)''')
connectionToDatabase.commit()
```

```
#Drop Transportation table if there already is one
cursor = connectionToDatabase.cursor()
cursor.execute('''drop table IF EXISTS Transportation''')
connectionToDatabase.commit()

#Create Transportation table
cursor.execute('''create table Transportation (TransportationID PRIMARY KEY, LeavingFacilityAddress, CustomerID FORIEGN KEY, Route, Cost, Barcode FORIEGN KEY)''')
connectionToDatabase.commit()
```

Which websites have the highest total sales?

This question was integral to our understanding of where most of our profits come from, without knowing this, we would practically be blind when attempting to reform and improve our companys operating standards and methods going forward. We also wanted to know which stores make the most amount of sales on our website for marketing reasons. 


```
cursor.execute('''SELECT TotalSales, Name FROM WebsiteInformation ORDER BY cast(TotalSales as Int) DESC;''')
for row in cursor.fetchall():
  print (row)
```

What are the different types of items?

In setting up this database-oriented centralized online shopping conglomerate, we had compiled and worked with a slew of companies, many of whom sell items of many different classifications and types. Because of this, it is imperative that we be able to extract this information specifically from our database, and especially, be able to identify which kinds of items sell the most, which could be part of potential business-related moves our company may make in the future



```
cursor.execute('''SELECT DISTINCT(ItemType) FROM Items;''')
for row in cursor.fetchall():
  print (row)
```

Which websites have the highest commission rate?

Since we work with companies directly, they each have their own separate contracts and thus their own, typically distinct commission rates. One company may start with a higher commission rate than another for a plethora of different reasons, they could be a more popular or well-known brand worldwide or be a brand that is trending up, whatever the case may be. This means, though, that due to any potential discrepancies (caused by unforeseeable or erratic factors) between our consumer-base and the global online shopping consumer-base, some companies may unjustly be getting higher commissions than others that are outperforming them in sales through our database. These are very important to identify, so as to lower the commission rates of underperforming companies and potentially, if necessary, raise those of overperforming companies. In this way, this question and its corresponding queries will be used by our business officials in conjunction with queries related to question 1


```
cursor.execute('''SELECT CommissionRate, Name FROM WebsiteInformation ORDER BY cast(CommissionRate as Int) DESC;''')
for row in cursor.fetchall():
  print (row)
```
